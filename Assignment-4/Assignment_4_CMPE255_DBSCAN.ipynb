```python
from pycaret.clustering import *
```

```python
from time import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

Load make_blobs data from datasets

```python
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.50, random_state=0)

dataset = pd.DataFrame(X)
```

Set up environment

```python
clustering = setup(dataset)
```

Create Model

```python
model = create_model('kmeans')
```

Assign the model to get results

```python
results = assign_model(model)
```

```python
results.head()
```

```python
results['Cluster'].unique()
```

Plot the data points to view clusters

```python
plot_model(model)
```

**DBSCAN**

```python
clustering_dbscan = setup(dataset)
```

creating DBSCAN model

```python
model_dbscan = create_model('dbscan')
```

Assigning model to get results

```python
results = assign_model(model_dbscan)
```

plot the model
