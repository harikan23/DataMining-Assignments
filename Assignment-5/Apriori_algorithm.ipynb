```python
transaction_data.head(5)
```

```python
transaction_details = []

for i in range(0, 1418):
  transaction_details.append([str(transaction_data.values[i,j]) for j in range(0, 49)])

print(transaction_details)
```

Generating association rules using apriori algorithm with minimum support 0.0045

```python
association_rules = apriori(transaction_details, min_support=0.0045, min_length=2)
```

List of rules

```python
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
association_results[1]
```

Test Data. (test items set)

```python
# test list Dark Chocolate Minis
test_items = ['Dark Chocolate Minis', 'Organic Pink Lemonade Bunny Fruit Snacks', 'Peach-Pear Sparkling Water']
```

Verifying the set of rules generated for the test item set

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    check =  all(item in items for item in test_items)
    if check is True:
      print("{} are frequent basket items purchased".format(test_items))
      break
```

Printing all the rules generated with support, confidence, lift

```python
for item in association_results:
    pair = item[0] 
    print(pair)
    items = [x for x in pair]
    print("Rule: ", items)
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
```
